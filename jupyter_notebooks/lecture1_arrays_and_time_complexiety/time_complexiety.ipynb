```python
# Big O Time Complexities Illustrated in Python

import time
import matplotlib.pyplot as plt
import numpy as np

# Define example functions for different time complexities

def constant_time(n):
    return 1

def logarithmic_time(n):
    steps = 0
    i = 1
    while i < n:
        steps += 1
        i *= 2
    return steps

def linear_time(n):
    return n

def linearithmic_time(n):
    steps = 0
    for i in range(n): n logn
        j = 1
        while j < n:
            steps += 1
            j *= 2
    return steps

def quadratic_time(n):
    return n * n

def cubic_time(n):
    return n * n * n

def exponential_time(n):
    return 2 ** n

# Map of complexities and functions
functions = {
    'O(1)': constant_time,
    'O(log n)': logarithmic_time,
    'O(n)': linear_time,
    'O(n log n)': linearithmic_time,
    'O(n^2)': quadratic_time,
    'O(n^3)': cubic_time,
    
}


# Timing utility
def time_function(func, n):
    start = time.time()
    func(n)
    end = time.time()
    return end - start

# Evaluate times
n_values = [2**i for i in range(1, 10)]
results = {label: [] for label in functions if label != 'O(2^n)'}

for n in n_values:
    for label, func in functions.items():
        if label == 'O(2^n)':
            continue  # Skip timing exponential function
        try:
            t = time_function(func, n)
        except RecursionError:
            t = None
        results[label].append(t)



# Plotting step count instead of actual time
step_results = {label: [] for label in functions}

for n in n_values:
    for label, func in functions.items():
        try:
            steps = func(n)
        except RecursionError:
            steps = None
        step_results[label].append(steps)

plt.figure(figsize=(12, 8))
for label, steps in step_results.items():
    plt.plot(n_values, steps, label=label)

plt.xlabel('Input size (n)')
plt.ylabel('Number of Steps')
plt.title('Big O Time Complexity Visualization (Step Count)')
plt.legend()
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
# Map of complexities and functions
def time_and_plot(functions):
    

    # Timing utility
    def time_function(func, n):
        start = time.time()
        func(n)
        end = time.time()
        return end - start

    # Evaluate times
    n_values = [2**i for i in range(5, 11)]
    results = {label: [] for label in functions}

    for n in n_values:
        for label, func in functions.items():
            try:
                t = time_function(func, n)
            except RecursionError:
                t = None
            results[label].append(t)

    # Plotting
    plt.figure(figsize=(12, 8))
    for label, times in results.items():
        plt.plot(n_values, times, label=label)

    plt.xlabel('Input size (n)')
    plt.ylabel('Time (seconds)')
    plt.title('Big O Time Complexity Visualization')
    plt.legend()
    plt.xscale('log')
    plt.yscale('log')
    plt.grid(True)
    plt.show()
```
